```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('arquivos/Amazon Echo 2 Reviews.csv')
display(df)
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df['User Verified'] = df['User Verified'].fillna('Not Verified')
df = df.dropna(subset=['Review Text'])
```

```python
df = df.drop(columns=['Review Useful Count', 'Declaration Text'])
display(df)
```

```python
df['Review Date'] = pd.to_datetime(df['Review Date'])
display(df)
```

```python
df['Rating'].value_counts()
```

```python
df['Rating'].mean()
```

```python
df['User Verified'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
df['Rating'].value_counts().plot(kind='bar')
plt.title('Distribuições de avaliações')
plt.show()
```

```python
analise_avaliacao = df[['Rating', 'Review Color']].groupby('Review Color').value_counts()
display(analise_avaliacao)
```

```python
analise_avaliacao_verificado = df[['Rating', 'User Verified']].groupby('User Verified').value_counts()
display(analise_avaliacao_verificado)
```

```python
analise_pct = df.groupby('Review Color')['Rating'].value_counts(normalize=True)
display(analise_pct)
```

```python
def sentimento(x):
    if x >= 4:
        return 'Positivo'
    elif x == 3:
        return 'Neutro'
    else:
        return 'Negativo'

df['Sentimento'] = df['Rating'].apply(sentimento)
display(df)
```

```python
df.info()
display(df)
```

```python
df = df.rename(columns={'User Verified': 'Usuário Verificado', 'Review Color': 'Cor do produto', 'Rating': 'Avaliação', 'Review Date': 'Data de Avaliação'})
display(df)
```

```python
df['Usuário Verificado'] = df['Usuário Verificado'].replace({'Verified Purchase': 'Compra Verificada', 'Not Verified': 'Não verificada',})

df['Cor do produto'] = df['Cor do produto'].replace({'Black': 'Preto', 'White': 'Branco'})

display(df)
```

```python
df.to_csv('arquivos/reviews_limpo.csv', index=False)
```
